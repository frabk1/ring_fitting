```python
import os, numpy as np, matplotlib.pyplot as plt, ehtim as eh
from ringfit.extraction import rbp_visualization_frames

im = eh.image.load_fits(os.path.expanduser("~/Downloads/test.fits"))
base = im.imarr()

frames, pts_by_panel, R, iters = rbp_visualization_frames(
    im,
    n_panels=5,
    max_it=5000,
    threshold=.75,
    radius=35.0
)

fig, axes = plt.subplots(1, 5, figsize=(19, 4), constrained_layout=True)
for ax, pts, it in zip(axes, pts_by_panel, iters):
    ax.imshow(base, origin="lower", cmap="inferno")
    if len(pts):
        ax.scatter(pts[:, 0], pts[:, 1], s=12, c="white")
        for x, y in pts:
            ax.add_patch(plt.Circle((x, y), R, facecolor="black", alpha=0.3, edgecolor="white", linewidth=1))
    ax.set_title(f"iteration={it}")
    ax.set_xticks([]); ax.set_yticks([])
plt.show()
```
